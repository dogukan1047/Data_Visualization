```python
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np
```

```python
ageList=[10,15,20,25,32,37,45,55,75]
weightList=[20,45,65,68,72,75,65,88,90]
weightList=np.array(weightList)
ageList=np.array(ageList)
```

```python
figure=plt.plot(ageList,weightList,"r")
plt.title("Kilo listesi")
plt.xlabel("Age")
plt.ylabel("Weight")
figure
```

```python
numpyArray=np.linspace(0,10,20)
numpyArray2=numpyArray**2
```

```python
plt.plot(numpyArray,numpyArray2,"g--")
plt.title("ln")
```

```python
plt.subplot(1,2,1)
plt.plot(numpyArray,numpyArray2,"r--")
plt.subplot(1,2,2)
plt.plot(numpyArray2,numpyArray,"g--")
```

```python
myFigure=plt.figure()
figureAxes=myFigure.add_axes([0.2,0.2,0.9,0.9])
figureAxes.plot(numpyArray,numpyArray2)
figureAxes.set_xlabel("X ekseni")
figureAxes.set_ylabel("Y ekseni")
```

```python
myFigure=plt.figure()
figureAxes1=myFigure.add_axes([0.2,0.2,0.9,0.9])
figureAxes2=myFigure.add_axes([0.33,0.5,0.3,0.3])
figureAxes1.plot(numpyArray,numpyArray2,"g--")
figureAxes1.set_xlabel("X ekseni")
figureAxes1.set_ylabel("Y ekseni")
figureAxes2.plot(numpyArray2,numpyArray,"r")
figureAxes2.set_xlabel("X ekseni")
figureAxes2.set_ylabel("Y ekseni")
```

```python
newFigure=plt.figure(dpi=100)
newAxis=newFigure.add_axes([0.1,0.1,0.9,0.9])
newAxis.plot(numpyArray,numpyArray2**2.5,label="numpyArray**2.5")
newAxis.plot(numpyArray,numpyArray2**3,label="numpyArray**3")
newAxis.legend()   
```
