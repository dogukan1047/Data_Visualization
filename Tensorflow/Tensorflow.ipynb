# Veriyi İkiye Bölme

````python
from sklearn.model_selection import train_test_split # Verileri test ve eğitim verilerine bölmek için gerekli
````

* y=wx +b
* y --> label
* x --> feature

````python
y_ekseni=myData["Fiyat"].values # np dizisine çevirdik
x_ekseni=myData[["BisikletOzellik1","BisikletOzellik2"]].values
````

````python
X_train, X_test, y_train, y_test = train_test_split(x_ekseni,y_ekseni,test_size=.33,random_state=15) 
# test ve eğitim verilerini böldük
````

``` Scaling ```

````python
from sklearn.preprocessing import MinMaxScaler # Verileri ölçekledik 
````

````python
scaler=MinMaxScaler()
scaler.fit(X_train)
MinMaxScaler(copy=True, feature_range=(0, 1))
````

````python
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
````

````python
import tensorflow as tf
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense
````

````python
model=Sequential()
model.add(Dense(4,activation="relu")) # Gizli katmanlarımıza nöron ekledik
model.add(Dense(4,activation="relu"))
model.add(Dense(4,activation="relu"))

model.add(Dense(1)) # Çıktı nöron sayısı
````

````python
model.compile(optimizer="rmsprop",loss="mse") # Modeli oluşturduk  eğitmeye hazır 
````

````python
model.fit(X_train,y_train,epochs=250)
````

````python
loss=model.history.history["loss"]
````

````python
sns.lineplot(x=range(len(loss)),y=loss)

````

````python
trainLoss=model.evaluate(X_train,y_train,verbose=0)
testLoss=model.evaluate(X_test,y_test,verbose=0)
print(trainLoss,testLoss)
````
